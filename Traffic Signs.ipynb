```python
data = []
for i in paths:
    img = cv2.imread(i)
    img = cv2.resize(img,(32,32))
    data.append(img)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(np.array(data),np.array(labels),stratify=np.array(labels),test_size=0.2)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
x_train
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train
```

```python
import keras
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Dense,Flatten
from keras.utils import np_utils
```

```python
y_train
```

```python
pd.DataFrame(y_train).value_counts()
```

```python
len(pd.DataFrame(y_train).value_counts())
```

```python
len(pd.DataFrame(y_test).value_counts())
```

```python
y_train_cat=np_utils.to_categorical(y_train,59)
```

```python
y_test_cat=np_utils.to_categorical(y_test,59)
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),padding="valid",activation="relu",input_shape=(32,32,3)))
model.add(MaxPooling2D(pool_size=(3,3)))

model.add(Conv2D(64,(3,3),padding="valid",activation="relu"))
model.add(MaxPooling2D(pool_size=(3,3)))


model.add(Flatten())
model.add(Dense(128,activation="relu",kernel_initializer="he_uniform"))
model.add(Dense(59,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
model.fit(x_train,y_train_cat,batch_size=10,epochs=10,validation_split=0.2)
```

```python
y_pred=model.predict(x_test)
y_pred_classes=[np.argmax(i) for i in y_pred] 
```

```python
model.evaluate(x_test,y_test_cat)
```
