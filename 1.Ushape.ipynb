##### Using he-uniform

```python
import tensorflow
import tensorflow.keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,BatchNormalization
model1_=Sequential()
model1_.add(Dense(7,activation="relu",input_dim=2, kernel_initializer='he_uniform'))
model1_.add(Dense(5,activation="relu"))
model1_.add(Dense(1,activation="sigmoid"))
model1_.add(BatchNormalization())
model1_.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
save_=model1_.fit(X_train,y_train,batch_size=10,epochs=150,verbose=1,validation_split=0.2)
```

```python
pred_1=model1_.predict(X_train)
pred_1=np.ravel(np.where(pred_1<0.5,0,1))
pred_1=pd.Series(pred_1)
pred_1.index=y_train.index
```

```python
import seaborn as sns
sns.scatterplot(X_train[0],X_train[1],hue=pred_1)
```

```python
from sklearn import metrics  #he-normal gives less accuracy comparitively with he-uniform
metrics.accuracy_score(y_train,pred_1)
```
