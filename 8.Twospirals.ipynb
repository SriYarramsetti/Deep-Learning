##### Using Batch Normalization

```python
import tensorflow
import tensorflow.keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
model8_=Sequential()
```

```python
model8_.add(Dense(6,activation="relu",input_dim=2, kernel_initializer='glorot_normal'))
model8_.add(Dense(3,activation="relu", kernel_initializer='glorot_normal'))
model8_.add(BatchNormalization())
model8_.add(Dense(1,activation="sigmoid"))
```

```python
model8_.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
model8_.fit(X_train,y_train,batch_size=10,epochs=200,verbose=1,validation_split=0.2)
```

```python
pred8=model8_.predict(X_train)
```

```python
pred8=np.ravel(np.where(pred8<0.5,0,1))
pred8=pd.Series(pred8)
pred8.index=y_train.index
pred8
```
