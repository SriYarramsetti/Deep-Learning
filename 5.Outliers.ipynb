```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data5[[0,1]],data5[2],test_size=0.15)
```

```python
from sklearn.preprocessing import StandardScaler
scaler5=StandardScaler()
```

```python
data_rescaled=scaler5.fit_transform(X_train)
```

```python
import tensorflow
import tensorflow.keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
model5=Sequential()
```

```python
model5.add(Dense(7,activation="sigmoid",input_dim=2))
model5.add(Dense(5,activation="sigmoid"))
model5.add(Dense(4,activation="sigmoid"))
model5.add(Dense(3,activation="sigmoid"))
model5.add(Dense(1,activation="sigmoid"))
```

```python
model5.summary()
```

```python
model5.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
model5.fit(X_train,y_train,batch_size=10,epochs=300,verbose=1)
```

```python
pred_=model5.predict(X_train)
```

```python
pred_=np.ravel(np.where(pred_<0.5,0,1))
pred_=pd.Series(pred_)
pred_.index=y_train.index
pred_
```
